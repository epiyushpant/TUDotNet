##  Data Access Technologies

In .NET, data access technologies are the frameworks and APIs that allow applications to interact with databases, XML, and other data sources. The most widely used today are Entity Framework Core (EF Core), ADO.NET, LINQ, and lightweight tools like Dapper.

| Technology | Description | Best Use Cases | Pros | Cons |
| --- | --- | --- | --- | --- |
| **ADO.NET** | Low-level API for direct database access using SQL commands, connections, and data readers. | High-performance apps needing fine-grained control over queries and transactions. | Very fast, mature, works with any database, full control. | Verbose, requires manual SQL and object mapping, more boilerplate code. |
| **Entity Framework Core (EF Core)** | Object-relational mapper (ORM) that maps .NET objects to database tables. | Modern enterprise apps, CRUD-heavy systems, cross-platform solutions. | Simplifies development, strong LINQ integration, reduces boilerplate, supports migrations. | Performance overhead compared to raw SQL, less control for complex queries. |
| **LINQ (Language Integrated Query)** | Query syntax integrated into C# for collections, XML, and databases. | Querying in-memory data, XML, or EF Core entities. | Readable, type-safe queries, consistent syntax across data sources. | Limited for highly optimized SQL scenarios, not a full ORM. |
| **Dapper** | Lightweight micro-ORM developed by Stack Overflow team. | Apps needing both performance and convenience, especially with custom SQL. | Extremely fast, minimal overhead, easy to use, simple mapping. | Lacks advanced ORM features (no change tracking, migrations). |

# ADO.NET Architecture in C#

## Introduction
ADO.NET (ActiveX Data Objects for .NET) is a data access technology provided by the .NET Framework to interact with relational databases. It enables communication between a C# application and data sources such as SQL Server, Oracle, and other databases.

## Key Components of ADO.NET
ADO.NET consists of two primary components:
1. Connected Architecture (Using Data Providers)
2. Disconnected Architecture (Using DataSet)

---

## 1. Connected Architecture (Data Provider Model)
A Data Provider is a set of components that facilitate interaction with a data source. The main components include:
- **Connection**: (SqlConnection, OleDbConnection) : Establishes a connection to the database.
- **Command**: (SqlCommand)  Executes SQL commands (SELECT, INSERT, UPDATE, DELETE).
- **DataReader**: (SqlDataReader) Provides fast, forward-only access to query results.
- **DataAdapter**: (SqlDataAdapter) Bridges the gap between the DataSet and the database.


### Example of Connected Architecture
```csharp
using System;
using System.Data.SqlClient;

class Program
{
    static void Main()
    {
        string connectionString = "your_connection_string";
        using (SqlConnection conn = new SqlConnection(connectionString))
        {
            conn.Open();
            string query = "SELECT * FROM Employees";
            SqlCommand cmd = new SqlCommand(query, conn);
            SqlDataReader reader = cmd.ExecuteReader();
            while (reader.Read())
            {
                Console.WriteLine(reader["Name"].ToString());
            }
        }
    }
}


# Disconnected Architecture (Using DataSet)

A **DataSet** is an in-memory representation of data that allows working with data offline.

## Key Components

- **DataSet**  
  Stores data in a disconnected mode.

- **DataTable**  
  Represents a table inside a DataSet.

- **DataRelation**  
  Defines relationships between DataTables.

- **DataAdapter**  
  Fills the DataSet with data from the database.


# Example: Disconnected Architecture with DataSet

This program demonstrates how to use a **DataSet** in disconnected mode to fetch and display data from a database.

```csharp
using System;
using System.Data;
using System.Data.SqlClient;

class Program
{
    static void Main()
    {
        string connectionString = "your_connection_string";

        using (SqlConnection conn = new SqlConnection(connectionString))
        {
            string query = "SELECT * FROM Employees";
            SqlDataAdapter adapter = new SqlDataAdapter(query, conn);

            // Create a DataSet to hold data in memory
            DataSet ds = new DataSet();

            // Fill DataSet with data from Employees table
            adapter.Fill(ds, "Employees");

            // Iterate through rows in the DataTable
            foreach (DataRow row in ds.Tables["Employees"].Rows)
            {
                Console.WriteLine(row["Name"].ToString());
            }
        }
    }
}


# Comparison of Connected and Disconnected Architecture

## Connected Architecture
- Directly in memory  
- Faster (requires open connection)  
- Forward-only access  
- Best for real-time transactions  

## Disconnected Architecture
- Uses DataSet (in-memory)  
- Slower due to caching  
- Supports navigation, updates, and caching  
- Best for large data processing  

---

# Advantages of ADO.NET

- **Scalability**: Supports both connected and disconnected models.  
- **Performance**: Optimized for high-performance data access.  
- **Interoperability**: Works with different databases like SQL Server, Oracle, MySQL.  
- **Security**: Uses connection pooling and authentication mechanisms.  


### Conclusion
ADO.NET provides a flexible and efficient way to access data in C# applications. Whether using the connected or disconnected approach, ADO.NET ensures effective database interaction while maintaining high performance and scalability.

![ADO.NET Architecture](ADOarchitecture.png)
